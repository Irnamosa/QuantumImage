# INCQI Implementation
From following [paper](https://www.nature.com/articles/s41598-021-93471-7#:~:text=Quantum%20image%20representation%20(QIR)%20is,role%20in%20quantum%20information%20processing.&text=The%20prepared%20image%20information%20was,quantum%20information%20was%20successfully%20realized.)

```python
from commons import *
from utils import *
import matplotlib.pyplot as plt
import math
import seaborn as sns
from qiskit.circuit.library.standard_gates import RYGate, XGate
from qiskit.visualization import plot_histogram
from qiskit.circuit import qpy_serialization
import pandas as pd
import numpy as np
from PIL import Image
```

```python
def counts2pix(counts):
    counts = [(k[:Nqs], k[Nqs:], v) for k,v in counts.items()]
    df = pd.DataFrame(counts, columns=["pos", "greyscale", "count"])
    df = df[(df.groupby(['pos'])['count'].transform(max) == df['count'])]
    df = df[~df.pos.duplicated(keep='first')]
    df = df.set_index('pos')
    missing = [k for k,v in populate(df['count'].to_dict(), Nqs).items() if v==0]
    df = df.drop('count', 1)
    df = df.applymap(lambda x: int(x,2))
    df_missing = pd.DataFrame(np.zeros(len(missing)), columns=['greyscale'], index=missing, dtype=int)
    df = df.append(df_missing)
    df = df.sort_index()
    return np.array(df.greyscale).reshape(N1,N2)
```

```python
img = Image.open("images/src/zero.png").convert("L")
display(img.resize((100, 100),  Image.NEAREST))
```

```python
mat = np.array(img)
```

```python
N1 = mat.shape[0]
N2 = mat.shape[1]
# Number of qubits required to sufficiently store the N1xN2 pixel positions 
Nqs = math.ceil(np.log2(N1*N2))
Nqs_gray = 8
```

```python
vbin = np.vectorize(lambda x: get_bin(x, Nqs_gray))
```

```python
# +8 qubits encoding grayscale values
qc = QuantumCircuit(Nqs+8,Nqs+8)

# Apply INCQI algorithm: https://doi.org/10.1038/s41598-021-93471-7
qc.h(range(0,Nqs))

bin_mat = vbin(mat)

for i in range(N1):
    for j in range(N2):
        for k,v in enumerate(bin_mat[i][j]):
            if(v=='1'):
                r = XGate()
                r = r.control(Nqs, ctrl_state=coord2bool(i,j,N1))
                qc.append(r,list(range(0,Nqs))+[Nqs+k])

qc.measure(np.arange(0,Nqs+8),np.arange(0,Nqs+8)[::-1]);
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=8192)
counts = job.result().get_counts()
plot_histogram(counts)
```

```python
mat_rec = counts2pix(counts)
```

```python
img_rec = Image.fromarray(np.uint8(mat_rec) , 'L')
display(img_rec.resize((100, 100),  Image.NEAREST))
```

## Noise Model: Thermal Relaxation

```python
from qiskit.providers.aer import AerSimulator
sim_toronto = AerSimulator.from_backend(provider.get_backend("ibmq_toronto"))
```

```python
import qiskit.providers.aer.noise as noise
from qiskit.providers.aer.noise import thermal_relaxation_error
from qiskit.providers.aer.noise import NoiseModel
```

```python
# From https://qiskit.org/documentation/tutorials/simulators/3_building_noise_models.html
# T1 and T2 values for qubits 0-3
nqs = 5

T1s = np.random.normal(100e3, 1e3, nqs) # Sampled from normal distribution mean 50 microsec
T2s = np.random.normal(5e3, 1e3, nqs)  # Sampled from normal distribution mean 50 microsec

# Truncate random T2s <= T1s
T2s = np.array([min(T2s[j], 2 * T1s[j]) for j in range(nqs)])

# Instruction times (in nanoseconds)
time_u1 = 0   # virtual gate
time_u2 = 50  # (single X90 pulse)
time_u3 = 100 # (two X90 pulses)
time_cx = 300
time_reset = 1000  # 1 microsecond
time_measure = 1000 # 1 microsecond

# QuantumError objects
errors_reset = [thermal_relaxation_error(t1, t2, time_reset)
                for t1, t2 in zip(T1s, T2s)]
errors_measure = [thermal_relaxation_error(t1, t2, time_measure)
                  for t1, t2 in zip(T1s, T2s)]
errors_u1  = [thermal_relaxation_error(t1, t2, time_u1)
              for t1, t2 in zip(T1s, T2s)]
errors_u2  = [thermal_relaxation_error(t1, t2, time_u2)
              for t1, t2 in zip(T1s, T2s)]
errors_u3  = [thermal_relaxation_error(t1, t2, time_u3)
              for t1, t2 in zip(T1s, T2s)]
errors_cx = [[thermal_relaxation_error(t1a, t2a, time_cx).expand(
             thermal_relaxation_error(t1b, t2b, time_cx))
              for t1a, t2a in zip(T1s, T2s)]
               for t1b, t2b in zip(T1s, T2s)]

# Add errors to noise model
noise_thermal = NoiseModel()
for j in range(nqs):
    noise_thermal.add_quantum_error(errors_reset[j], "reset", [j])
    noise_thermal.add_quantum_error(errors_measure[j], "measure", [j])
    noise_thermal.add_quantum_error(errors_u1[j], "u1", [j])
    noise_thermal.add_quantum_error(errors_u2[j], "u2", [j])
    noise_thermal.add_quantum_error(errors_u3[j], "u3", [j])
    for k in range(nqs):
        noise_thermal.add_quantum_error(errors_cx[j][k], "cx", [j, k])
```

```python
backend = provider.get_backend('ibmq_toronto')
```

```python
# Get coupling map from backend
coupling_map = backend.configuration().coupling_map

# Get basis gates from noise model
basis_gates = NoiseModel.from_backend(backend).basis_gates
```

```python
job_sim = execute(qc, Aer.get_backend('qasm_simulator'),
         coupling_map=coupling_map,
         basis_gates=basis_gates,
         noise_model=noise_thermal, shots=200)
```

```python
counts = job_sim.result().get_counts(qc)
```

```python
mat_rec = counts2pix(counts)
```

```python
img_rec = Image.fromarray(np.uint8(mat_rec) , 'L')
display(img_rec.resize((100, 100),  Image.NEAREST))
```

## Truncated method

```python
unique_vals = np.unique(mat.reshape(16))
discretise_map = {k:v for k,v in zip(unique_vals, range(0,len(unique_vals)))}
discretise_map_rev = dict((v,k) for k,v in discretise_map.items())
```

```python
discretise = lambda x: discretise_map[x]
discretise_rev = lambda x: discretise_map_rev[x]
vdiscretise = np.vectorize(discretise)
vdiscretise_rev = np.vectorize(discretise_rev)
```

```python
mat_disc = vdiscretise(mat)
```

```python
Nqs_gray = len(get_bin(np.max(mat_disc),0))
```

```python
# +2 qubits encoding grayscale values
qc = QuantumCircuit(Nqs+Nqs_gray,Nqs+Nqs_gray)

# Apply INCQI algorithm: https://doi.org/10.1038/s41598-021-93471-7
qc.h(range(0,Nqs))

bin_mat = vbin(mat_disc)

for i in range(N1):
    for j in range(N2):
        #qc_t =  QuantumCircuit(1, name=str(N*i+j))
        #qc_t.ry(angles[i][j],0)
        #custom = qc_t.to_gate().control(Nqs, ctrl_state=coord2bool(i,j,N))

        for k,v in enumerate(bin_mat[i][j]):
            if(v=='1'):
                r = XGate()
                r = r.control(Nqs, ctrl_state=coord2bool(i,j,N1))
                qc.append(r,list(range(0,Nqs))+[Nqs+k])

qc.measure(np.arange(0,Nqs+Nqs_gray),np.arange(0,Nqs+Nqs_gray)[::-1]);
```

```python
qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=8192)
counts = job.result().get_counts()
plot_histogram(populate(counts, Nqs+Nqs_gray))
```

```python
mat_rec = counts2pix(counts)
mat_rec = vdiscretise_rev(mat_rec)
img_rec = Image.fromarray(np.uint8(mat_rec) , 'L')
display(img_rec.resize((100, 100),  Image.NEAREST))
```
